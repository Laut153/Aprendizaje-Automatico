```python
df= pd.read_csv('/content/drive/MyDrive/Inteligencia Artificial/segundo año/primer semestre/aprendizaje automatico/Aprendizaje_Automatico_1/set_datos/winequality-white.csv', sep=';')
df
```

```python
df.quality.unique()
```

```python
x= df.drop(["quality"], axis=1)
x
```

```python
y= df.quality
y
```

```python
train_x,test_x,train_y,test_y= train_test_split(x,y, test_size=0.2,random_state=123)
```

#Resolucion del problema usando modelo de arbol de decision

```python
modelo=DecisionTreeClassifier(random_state=123, max_depth=4)
modelo.fit(train_x,train_y)
pred_y=modelo.predict(test_x)
from sklearn.metrics import accuracy_score
acc= accuracy_score(test_y,pred_y)
acc
```

# Resolucion del problema usando modelo de KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=20,weights="distance")
knn.fit(train_x,train_y)
predknn_y=knn.predict(test_x)
ac= accuracy_score(test_y,predknn_y)
ac
```
